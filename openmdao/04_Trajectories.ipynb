# Multirotor trajectory and design optimization with FAST-OAD

FAST-OAD is a framework for performing rapid Overall Aircraft Design co-developped by ONERA and ISAE-SUPAERO. The computational core of FAST-OAD is based on the  [OpenMDAO framework](https://openmdao.org/).

## 1. Setting up and analyzing the initial problem

To organize our work, we propose to use two user folders `data/` and `workdir/`. For instance, in `data/` we store a XML file which describes the the multirotor. In `workdir/`, we store files generated or modified by FAST-OAD.

```python
import numpy as np

import matplotlib.pyplot as plt

plt.rc("font", family="serif")
plt.rc("xtick", labelsize="small")
plt.rc("ytick", labelsize="small")

res = optim_problem_4.model.trajectory.trajectory.res
problem = optim_problem_4

t = np.array(res["time"])
xpp = res["droneMassPropeller.der(xp)"]
xp = res["droneMassPropeller.xp"]
x = res["droneMassPropeller.x"]
n = res["droneMassPropeller.n"]
T = res["droneMassPropeller.T"]
power = res["droneMassPropeller.Power"]

if problem:
    t0 = problem["data:trajectory:time"]
    T0 = problem["data:trajectory:torque"]


fig, ((ax1, ax2), (ax3, ax4), (ax5, ax6)) = plt.subplots(
    3, 2, sharex="col", figsize=(10, 6)
)

# Acceleration
l1 = ax1.plot(t, xpp)
ax1.set_ylabel("Acceleration [m/s²]")
# ax1.set_xlabel("Time [s]")
ax1.grid()

# Speed
ax3.plot(t, xp)
ax3.set_ylabel("Speed [m/s]")
# ax2.set_xlabel("Time [s]")
ax3.grid()

# Position
ax5.plot(t, x)
ax5.set_ylabel("Position [m]")
ax5.set_xlabel("Time [s]")
ax5.grid()

# Motor speed
ax2.plot(t, n)
ax2.set_ylabel("Motor rotational speed [Hz]")
# ax4.set_xlabel("Time [s]")
ax2.grid()

# Motor torque
ax4.plot(t, T)
if problem:
    ax4.scatter(t0, T0, color='k', label="Control points")
    ax4.scatter(t0, T0)
    ax4.legend()
ax4.set_ylabel("Motor torque [N.m]")
# ax5.set_xlabel("Time [s]")
ax4.grid()

# Total electric  power
ax6.plot(t, power * 1e-3)
ax6.set_ylabel("Total electric power [kW]")
ax6.set_xlabel("Time [s]")
ax6.grid()

res = optim_problem_8.model.trajectory.trajectory.res
problem = optim_problem_8

t = np.array(res["time"])
xpp = res["droneMassPropeller.der(xp)"]
xp = res["droneMassPropeller.xp"]
x = res["droneMassPropeller.x"]
n = res["droneMassPropeller.n"]
T = res["droneMassPropeller.T"]
power = res["droneMassPropeller.Power"]

if problem:
    t0 = problem["data:trajectory:time"]
    T0 = problem["data:trajectory:torque"]


# Acceleration
l2 = ax1.plot(t, xpp)

# Speed
ax3.plot(t, xp)

# Position
ax5.plot(t, x)

# Motor speed
ax2.plot(t, n)

# Motor torque
ax4.plot(t, T)
if problem:
    ax4.scatter(t0, T0)

# Total electric  power
ax6.plot(t, power * 1e-3)


res = optim_problem_16.model.trajectory.trajectory.res
problem = optim_problem_16

t = np.array(res["time"])
xpp = res["droneMassPropeller.der(xp)"]
xp = res["droneMassPropeller.xp"]
x = res["droneMassPropeller.x"]
n = res["droneMassPropeller.n"]
T = res["droneMassPropeller.T"]
power = res["droneMassPropeller.Power"]

if problem:
    t0 = problem["data:trajectory:time"]
    T0 = problem["data:trajectory:torque"]


# Acceleration
l3 = ax1.plot(t, xpp)

# Speed
ax3.plot(t, xp)

# Position
ax5.plot(t, x)

# Motor speed
ax2.plot(t, n)

# Motor torque
ax4.plot(t, T)
if problem:
    ax4.scatter(t0, T0)

# Total electric  power
ax6.plot(t, power * 1e-3)

res = optim_problem_32.model.trajectory.trajectory.res
problem = optim_problem_32

t = np.array(res["time"])
xpp = res["droneMassPropeller.der(xp)"]
xp = res["droneMassPropeller.xp"]
x = res["droneMassPropeller.x"]
n = res["droneMassPropeller.n"]
T = res["droneMassPropeller.T"]
power = res["droneMassPropeller.Power"]

if problem:
    t0 = problem["data:trajectory:time"]
    T0 = problem["data:trajectory:torque"]


# Acceleration
l4 = ax1.plot(t, xpp)

# Speed
ax3.plot(t, xp)

# Position
ax5.plot(t, x)

# Motor speed
ax2.plot(t, n)

# Motor torque
ax4.plot(t, T)
if problem:
    ax4.scatter(t0, T0)

# Total electric  power
ax6.plot(t, power * 1e-3)

line_labels = ["N=4", "N=8", "N=16", "N=32"]

# Create the legend
legend = fig.legend([l1, l2, l3, l4],     # The line objects
           labels=line_labels,   # The labels for each line
           loc="lower center",   # Position of legend
           # borderaxespad=0.1,    # Small spacing around legend box
           title="Number of control points",  # Title for the legend
           bbox_to_anchor=(0.5, 1.01),
           ncol=4, fancybox=True, shadow=True
           )

fig.align_ylabels()
fig.tight_layout()
# plt.legend()
fig.savefig("trajectories_control_points.pdf", bbox_extra_artists=(legend,), bbox_inches='tight')

plt.show()
```
